## Reading the Dataset <a id='Reading-the-Dataset'></a>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Migration.csv')
```

```python
df.set_index('Unnamed: 0', inplace=True)
df.index.name = None
```

```python
years = list(map(str, range(1980, 2014)))
years[:5]
```

## Visualizing the data using Matplotlib <a id='two'></a>

```python
mpl.style.use('ggplot')
```

## Pie Chart <a id="piechart"></a>

```python
tot = df[["Continent","total"]].groupby("Continent").sum()
tot
```

```python
tot['total'].plot(kind='pie',
                  figsize=(10,5),
                  autopct='%1.1f%%',
                  startangle=90,
                  shadow=True,
                  pctdistance=1.14,
                  labels=None,
                  explode=[0.1,0,0,0,0.1,0.1])

plt.axis('equal');
plt.legend(labels=tot.index, loc='upper left', fontsize=7)
plt.title("Immigration to Canada by Continent (1980-2013)", y=1.1, fontsize=11)
```

```python
tot = df[['Continent','2013']].groupby('Continent').sum()
tot['2013'].plot(kind='pie',
                 figsize=(10,5),
                 shadow=True,
                 autopct='%1.1f%%',
                 pctdistance=1.2,
                 labels=None)

plt.legend(labels = tot.index, loc='upper left', fontsize=8)
plt.axis('equal');
plt.title("Immigration to Canada in 2013 of Continents");
```

## Box Plot <a id="boxplot"></a>

```python
df.loc[['Japan'], years].T.plot(kind='box')
plt.title("Distribution of Immigrants from Japan");
plt.ylabel("No. of Immigrants")
```

```python
df.loc[['Japan'], years].T.describe().round(2)
```

Compare the distribution of the number of new immigrants from India and China for the period 1980 - 2013

```python
df_x = df.loc[['India','China'], years].T
df_x.plot(kind='box',
         figsize=(10,6),
         color='blue')
plt.title("Distribution of Immigrants from India and China");
plt.ylabel("No. of Immigrants")
plt.xlabel("Country")
```

## Sub Plot <a id="subplots"></a>

```python
fig = plt.figure()

ax0 = fig.add_subplot(1, 2, 1)
ax1 = fig.add_subplot(1, 2, 2)

df_x.plot(kind='box', ax=ax0, figsize=(20,7))
ax0.set_xlabel("Country")
ax0.set_ylabel("No. of Immigrants")

df_x.plot(kind='line', ax=ax1, figsize=(20,7))
ax1.set_xlabel("Year")
ax1.set_ylabel("No. of Immigrants")
```

```python
df2 = df[years].T
df2.index = df2.index.map(int)
df2 = df2
```

```python
def dec(row, year):
    tot = 0
    for i in range(year, year+10):
       tot += row[i]
    return tot

dec1 = df2.apply(dec, year=1980)
dec2 = df2.apply(dec, year=1990)
dec3 = df2.apply(dec, year=2000)
```

```python
decade = pd.DataFrame([dec1, dec2, dec3]).rename({0:"1980s", 1:"1990s", 2:"2000s"}).T
```

```python
dec_df = decade.sort_values(by=["1980s","1990s","2000s"], ascending=False).head(15)
```

```python
dec_df.plot(kind='box',
           figsize=(10,6),
           color='maroon')

plt.title("Immigration from Top 15 countries for decades 80s, 90s, and 2000s");
plt.ylabel("No. of Immigrants");
```

```python
dec_df.describe().round(2)
```

## Scatter Plot <a id="scatterplot"></a>

```python
df_tot = pd.DataFrame(df[years].sum())
df_tot.index = df_tot.index.map(int)
df_tot.reset_index(inplace=True)
df_tot.rename(columns={'index':'year', 0:"Total"}, inplace = True)
```

```python
df_tot.plot(kind='scatter',
            x="year",
            y="Total",
            color="maroon",
            alpha=0.8,
            figsize=(10,6));
plt.title("Total Migration to canada from 1980-2013", fontsize=15);
```

```python
x = df_tot['year']
y = df_tot['Total']
fit = np.polyfit(df_tot['year'], df_tot['Total'], deg=1)
fit
```

```python
df_tot.plot(kind='scatter',
            x="year",
            y="Total",
            color="darkblue",
            alpha=0.8,
            figsize=(10,6));
plt.title("Total Migration to canada from 1980-2013", fontsize=15);
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Immigrants', fontsize=14)

plt.plot(x, fit[0] * x + fit[1], color='red')
plt.annotate('y={0:.0f}*x + {1:.0f}'.format(fit[0], fit[1]), xy=(2003,200000), fontsize=13)
```

```python
df_countries = df.loc[['Denmark','Norway','Sweden'], years]
df_countries = pd.DataFrame({'Total': df_countries.sum()})
df_countries.index = df_countries.index.map(int)
df_countries.reset_index(inplace=True)
df_countries.head()
```

```python
df_countries.plot(kind='scatter',
                  x = 'index',
                  y = 'Total',
                  figsize=(10,6),
                  color = '#123456')

plt.title("Total Migration to canada from 1980-2013 (Denmark, Norway and Sweden)", fontsize=12);
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Immigrants', fontsize=14)
```

## Bubble Plot <a id="bubbleplot"></a>

> A bubble plot is a variation of the scatter plot that displays three dimensions of data (x, y, z). The data points are replaced with bubbles, and the size of the bubble is determined by the third variable z, also known as the weight.

```python
df_loc = df.loc[['Argentina','Brazil'], years].T
df_loc.index.name = "Year"
df_loc.index = df_loc.index.map(int)
df_loc.reset_index(inplace=True)
df_loc.head(2)
```

```python
df_loc.plot(kind='line',
            x='Year',
            lw=1.5);
```

Normalizing the values in the column brazil andargentina to bring the values in the range of (0, 1).
There are many formula for normalization. The general formula we'll use now is $X$ = 

**$X$ - $X$($min$)** <br>
    $-------$ <br>
     **$X$($max$) - $X$($min$)**

where $X$ is the original value, $X'$ is the corresponding normalized value. The formula sets the max value in the dataset to 1, and sets the min value to 0. The rest of the data points are scaled to a value between 0-1 accordingly.

```python
norm_Brazil = (df_loc.Brazil - df_loc.Brazil.min())/(df_loc.Brazil.max() - df_loc.Brazil.min())
norm_Argentina = (df_loc.Argentina - df_loc.Argentina.min())/(df_loc.Argentina.max() - df_loc.Argentina.min())
```

```python
# Let's plot the Bubble plot

ax0 = df_loc.plot(kind='scatter',
            x = 'Year',
            y = 'Brazil',
            c = 'green',
            figsize=(14,8),
            alpha=0.6,
            s = norm_Brazil*2500,
            xlim = (1975,2015),
            label="Brazil")

ax1 = df_loc.plot(kind = 'scatter',
            x = 'Year',
            y = 'Argentina',
            c = "red",
            alpha=0.5,
            ax = ax0,
            s = norm_Argentina*2500,
            label="Argentina")

ax0.set_ylabel("Number of Immigrants")
ax0.set_xlabel("Immigration from Brazil and Argentina")
ax0.legend(loc = 'upper left')
```

```python
norm_india = (df_x.India - df_x.India.min())/(df_x.India.max() - df_x.India.min())
norm_china = (df_x.China - df_x.China.min())/(df_x.China.max() - df_x.China.min())
```

```python
df_x.index = df_x.index.map(int)
df_x = df_x.reset_index().rename({'index':'Year'}, axis=1)
```

```python
ax0 = df_x.plot(kind='scatter',
                x = 'Year',
                y = 'India',
                alpha=0.5,
                figsize=(10,6),
                s = norm_india*850 + 10,
                xlim = (1975, 2015))

ax1 = df_x.plot(kind='scatter',
                x = 'Year',
                y = 'China',
                alpha = 0.5,
                ax = ax0, 
                c = "red",
                s = norm_china*850 + 10)

ax0.set_title("Immigrants from China and India from 1980 to 2013")
ax0.legend(["India","China"], fontsize=14, loc="upper left");
ax0.set_ylabel("No. of Immigrants", fontsize=12)
```
