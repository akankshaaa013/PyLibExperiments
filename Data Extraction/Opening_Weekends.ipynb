```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
```

```python
rank = []
release = []
opening = []
total_gross = []
total = []
theaters = []
avg = []
date = []
distributor = []
```

```python
for i in range(0, 801, 200):
    url = f"https://www.boxofficemojo.com/chart/top_opening_weekend/?ref_=bo_csw_ac&offset={i}"

    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')
    tables = soup.find_all('table')
    table = tables[0]

    rows = table.find_all('tr')

    for row in rows[1:]:
        cols = row.find_all('td')
        rank.append(cols[0].text.strip())
        release.append(cols[1].text.strip())

        opening_text = cols[2].text.strip().replace('$', '').replace(',', '')
        opening.append(opening_text if opening_text else "N/A")

        total_gross_text = cols[3].text.strip().replace('$', '').replace(',', '')
        total_gross.append(total_gross_text if total_gross_text else "N/A")

        total_percentage_text = cols[4].text.strip().replace('%', '')
        total.append(total_percentage_text if total_percentage_text else "N/A")

        theaters_text = cols[5].text.strip().replace(',', '')
        theaters.append(theaters_text if theaters_text else "N/A")

        avg_text = cols[6].text.strip().replace('$', '').replace(',', '')
        avg.append(avg_text if avg_text else "N/A")

        date_text = cols[7].text.strip()
        if date_text:
            date_obj = datetime.strptime(date_text, "%b %d, %Y")
            date.append(date_obj)
        else:
            date.append("N/A")

        distributor_text = cols[8].text.strip()
        # If distributor_text contains a link, extract only the text before the link
        if '(' in distributor_text:
            distributor_name = distributor_text.split(' (')[0]
        else:
            distributor_name = distributor_text
        distributor.append(distributor_name)
```

```python
print("Length of rank:", len(rank))
print("Length of release:", len(release))
print("Length of opening:", len(opening))
print("Length of total_gross:", len(total_gross))
print("Length of total:", len(total))
print("Length of theaters:", len(theaters))
print("Length of avg:", len(avg))
print("Length of date:", len(date))
print("Length of distributor:", len(distributor))
```

```python
data = {"Rank" : rank, "Release" : release, "Opening" : opening, "Total Gross" : total_gross, "% of Total" : total, "Theaters" : theaters, "Average" : avg, "Date" : date, "Distributor" : distributor}
df = pd.DataFrame(data)
```
