```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
rank = []
title = []
year = []
lifetime_gross = []
```

```python
for i in range(0, 10001, 200):
    url = f"https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW&offset={i}"
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
        
    tables = soup.find_all('table')
    table = tables[0]
        
    rows = table.find_all('tr')
        
    for row in rows[1:]:
        cols = row.find_all('td')
        rank.append(cols[0].text.strip())
        title.append(cols[1].text.strip())  # corrected this line
        year_text = cols[3].text.strip()
        years_part = year_text.split()
        for part in years_part:
            if part.isdigit() and len(part) == 4:
                year.append(part)
                break  # add break here to stop appending multiple years for a single movie
            else:
                year.append("N/A")
        
        gross_text = cols[2].text.strip().replace(',', '').replace('$', '')
        if gross_text:  # check if gross_text is not empty before appending
            lifetime_gross.append(gross_text)
        else:
            lifetime_gross.append("N/A")
```

```python
data = {'Rank': rank, 'Title': title, 'Year': year, 'Lifetime Gross': lifetime_gross}
df = pd.DataFrame(data)
```

```python
df['Year'] = pd.to_numeric(df['Year'])
df['Lifetime Gross'] = pd.to_numeric(df['Lifetime Gross'])
```

```python
df.to_csv("Top Highest Gross.csv")
```
